```python
from math import pi

ivals = list(range(9))
print(ivals)

pilist = [-pi + i*2*pi/8 for i in ivals]
print(pilist)
```

#### More compact version:

```python
from math import pi

pilist = [-pi + i*2*pi/8 for i in range(9)]

print(pilist)
```

* Note the `list()` function is not needed when only iterating over the values generated by `range()`
